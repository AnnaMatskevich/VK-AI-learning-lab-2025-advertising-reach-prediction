```python
def tz_loss_function(y_true, y_pred):
  eps = 0.005
  return 100 * np.exp(np.sum(np.abs(np.log((y_pred + eps) / (y_true + eps)))) / len(y_true)) - 100
```

```python
def calculate_modified_cdf(data):
    sorted_data = np.sort(np.array(data))
    unique_values, counts = np.unique(sorted_data, return_counts=True)
    result_probs = []
    values = []
    for i, value in enumerate(unique_values):
        less_than_prob = np.sum(sorted_data < value) / len(sorted_data)
        equal_to_prob = counts[i] / len(sorted_data)
        modified_prob = less_than_prob + 0.5 * equal_to_prob
        result_probs.append(modified_prob)
        values.append(value)

    return values, result_probs
```

```python
def calculate_user_sum_probability_2(row):
  """
  тоже самое что и в 1, но если у user count < 20. его не добавляем в сумму. потом сумму масштабируем
  типо если было 20 user, у 5и count < 20, о для остальных 15и посчитали ответ и домножили на 20/15

  """
  STATS_DURATION = history['hour'].max() - history['hour'].min()
  eps = 1e-6
  mask = users[users['user_id'].isin(row['user_ids'])]['count'] > 20
  p_win_vector = st.norm.cdf(np.array([np.log(row['cpm'])] * row['audience_size']),
                     users[users['user_id'].isin(row['user_ids'])]['mean'] + eps,
                     users[users['user_id'].isin(row['user_ids'])]['std'] + eps)[mask]



  count_auctions = np.array([(users.iloc[user_]['count'] / STATS_DURATION * row['duratin_hours']) for user_ in row['user_ids']])[mask]
  return np.sum(np.nan_to_num(1 - (1 - p_win_vector)**count_auctions)) * row['audience_size'] / np.sum(mask)
```

```python
import scipy.stats as st

def calculate_user_sum_probability_1(row, k1, k2):
  """
  Для каждого user в списке считаем вероятность выиграть, число аукционов. Дальше по формуле (1 - (1 - p_win)**count_auctions) для каждого user
  По user суммируем

  Как считаемся вероятность победы для каждого user:
    в users есть mean и std для каждого user. Считаем st.norm.cdf(cpm, mean, std)
  Как считается число аукционов для каждого пользователя:
    (число аукционов у этого user во всей history) / (history.hour.max() - history.hour.min()) * длительность показа рекламы
  """
  time_part = 0
  for i in range(row['hour_start'], row['hour_end'] + 1):
        time_part += hours24.loc[i % 24, 'count'] / len(history)
  STATS_DURATION = history['hour'].max() - history['hour'].min()
  eps = 1e-6
  p_win_vector = st.norm.cdf(np.array([np.log(row['cpm'])] * row['audience_size']),
                     users[users['user_id'].isin(row['user_ids'])]['mean'] + eps,
                     users[users['user_id'].isin(row['user_ids'])]['std'] + eps) * k1# может тут без копирования можно не писать дважды users[users['user_id'] in row['user_ids']]?

  count_auctions = np.array([(users.iloc[user_]['count'] / STATS_DURATION * row['duration_hours']) for user_ in row['user_ids']]) * k2
  return np.sum(np.nan_to_num(1 - (1 - p_win_vector)**count_auctions))
```

```python
import scipy.stats as st
from collections import Counter

def calculate_user_sum_probability_3(row, k1, k2):
  ans = 0
  # coeff = 0
  # for h in range(row['hour_start'], row['hour_end'] + 1):
  #   coeff += hours24['count'].iloc[h%24]
  # coeff /= (row['duration_hours'] * hours24['count'].mean())
  counter = Counter([h%24 for h in range(row['hour_start'], row['hour_end'] + 1)])
  eps = 1e-8
  STATS_DURATION = history['hour'].max() - history['hour'].min()
  for user in row['user_ids']:
    user_ans = 1
    for pub_ in row['publishers']:
      for hour in counter.keys():
        if (user, pub_, hour) not in user_publisher_hour:
          continue
        arr = user_publisher_hour[(user, pub_, hour)]
        mean, std, count = arr[0], arr[1], arr[2]
        if count == 0:
          continue
        if std == 0:
          std = eps
        p_win = st.norm.cdf(np.log(row['cpm']), mean, std) * k1
        n_aucions = count / STATS_DURATION * row['duration_hours']  * k2 #* coeff
        user_ans *= (1 - (1 - p_win)**n_aucions)**counter[hour]
    ans += user_ans
  return ans
```

```python
import scipy.stats as st
from collections import Counter

def calculate_classes_sum_probability(row, k1, k2):
  """
  лучше даже не вникать
  """
  ans = 0
  coeff = 0
  for h in range(row['hour_start'], row['hour_end'] + 1):
    coeff += hours24['count'].iloc[h%24]
  coeff /= (row['duration_hours'] * hours24['count'].mean())
  STATS_DURATION = history['hour'].max() - history['hour'].min()

  classes = [users['class'].iloc[u] for u in row['user_ids']]

  counter = Counter(classes)
  for class_ in counter.keys():
    cpms = np.array([])
    for pub_ in row['publishers']:
      cpms = np.concatenate((cpms, classes_data[f'publisher_{int(pub_)}'].iloc[class_]))
    mean = 0
    std = 0
    if len(cpms) == 0:
      continue
    elif len(cpms) == 1:
      mean = np.mean(cpms)
      std = 1e-7
    else:
      mean = np.mean(cpms)
      std = max(np.std(cpms), 1e-7)
    p_win = st.norm.cdf(np.log(row['cpm']), mean, std) * k1
    if np.isnan(p_win):
      print(cpms, mean, std)
    n_aucions = len(cpms) / STATS_DURATION * row['duration_hours'] * coeff * k2
    ans += (1 - (1 - p_win)**n_aucions) * counter[class_]
  return ans
```
